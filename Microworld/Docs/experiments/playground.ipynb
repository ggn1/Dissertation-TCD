This file analyzes data from simulation runs.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
land_size = {"rows": 6, "columns": 6}
time_max = 300
```

```python
df = pd.read_csv("data/all_coniferous_0to300.csv")
df
```

```python
time_slice_range = [70, 90]
df_slice = df[np.logical_and(
    df.time >= time_slice_range[0], 
    df.time <= time_slice_range[1]
)]
plt_figsize = (12, 6)
```

```python
# STRESS V/S TIME
plt.figure(figsize=plt_figsize)
for r in range(land_size['rows']):
    for c in range(land_size['columns']):
        plt.plot(df_slice['time'], df_slice[f'p{r}{c}_stress'], label=f'p{r}{c}')
plt.xlabel("Year")
plt.ylabel("Stress")
plt.yscale("symlog")
plt.title("Stress over Time")
plt.show()
```

```python
# CO2 V/S TIME
plt.figure(figsize=plt_figsize)
plt.plot(df_slice['time'], df_slice['co2_ppm'])
plt.xlabel("Year")
plt.ylabel("CO2 (ppm)")
plt.yscale("linear")
plt.title("Atmospheric CO2 Concentration over Time")
plt.show()
```

```python
# TREE LIFE STAGE V/S TIME
plt.figure(figsize=plt_figsize)
for r in range(land_size['rows']):
    for c in range(land_size['columns']):
        plt.plot(df_slice['time'], df_slice[f'p{r}{c}_stage'], label=f'p{r}{c}')
plt.xlabel("Year")
plt.ylabel("Life Stage")
plt.yscale("linear")
plt.title("Tree Life Stage over Time")
plt.show()
```
