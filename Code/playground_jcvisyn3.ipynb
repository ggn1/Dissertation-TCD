```python
useful_features = [
    "gene_name_short", "gene_name_long", "locus_tag", 
    "description", "plasmid_size", "construct_insert_size",
    "production_sequence", "documentation_image"
]

df_u = df[useful_features]
desc_df(df_u)
```

```python
CODON_TABLE = pd.read_csv("../__Data/codon_table.csv")
CODON_TABLE = CODON_TABLE[CODON_TABLE['amino_acid_code']!="START"]
CODON_TABLE = CODON_TABLE.set_index('dna_codon', drop=True).to_dict()['amino_acid_code']

def nucleotide_to_amino_acid(nuc_seq):
    ''' Converts nucleotide sequence into IUPAC amino acid sequence.
        @param nuc_seq: Nucleotide sequence.
    '''
    aa_seq = ""
    
    while len(nuc_seq) > 0:
        codon = nuc_seq[0:3]
        amino_acid = CODON_TABLE[codon]
        nuc_seq = nuc_seq[3:]
        if not "<" in aa_seq:
            if amino_acid == "M": aa_seq += "<"
            else: aa_seq += codon
        elif amino_acid == "STOP":
            aa_seq += ">" + nuc_seq
            break
        else: aa_seq += amino_acid
    
    return aa_seq
```

```python
# Express nucleotide sequences as amino acid sequences.
aa_seqs = df_u.apply(lambda row: nucleotide_to_amino_acid(row.production_sequence), axis=1)
df_u = df_u.assign(amino_acid_sequence=aa_seqs)
df_u
```

```python
# Do all amino acid sequences have a start codon?
has_start = True
for i, row in df_u.iterrows():
    aa_seq = row['amino_acid_sequence']
    if '<' not in aa_seq: 
        has_start = False
        break
print(has_start)
```

```python
# Do all amino acid sequences have a stop codon?
has_stop = True
for i, row in df_u.iterrows():
    aa_seq = row['amino_acid_sequence']
    if '>' not in aa_seq: 
        has_stop = False
        break
print(has_stop)
```

```python
aa_seq_len = df_u['amino_acid_sequence'].apply(lambda aa_seq: aa_seq[aa_seq.find('<')+1:aa_seq.find('>')]).apply(len)
print(min(aa_seq_len), max(aa_seq_len))
```
